```python
import numpy as np
import matplotlib.pyplot as plt

time = np.linspace(0, 10, 101)
print("Time = ", time)
```

```python
B, A, alpha, beta = 0, 1, 0.5, 1.5
viral_load = A * np.exp(-alpha*time) + B * np.exp(-beta*time)
plt.plot(time, viral_load)
plt.title("B = 0, A = 1, Alpha = 0.5, Beta = 1.5", weight='bold')
plt.xlabel('t', weight='bold')
plt.ylabel('V(t)', weight='bold')
plt.show()
```

```python
fig, axs = plt.subplots(nrows=4, ncols=1, sharex=True)

viral_load = A * np.exp(-alpha*time) + B * np.exp(-beta*time)

viral_load = 1 * np.exp(-0.5*time) + 100 * np.exp(-1.5*time)
axs[0].plot(time, viral_load)

viral_load = 20 * np.exp(-0.5*time) + 0 * np.exp(-1.5*time)
axs[1].plot(time, viral_load)

viral_load = 1 * np.exp(-10*time) + 0 * np.exp(-1.5*time)
axs[2].plot(time, viral_load)

viral_load = 1 * np.exp(-0.5*time) + 0 * np.exp(-30*time)
axs[3].plot(time, viral_load)

axs[0].set_title("B = 100, A = 1, Alpha = 0.5, Beta = 1.5")
axs[1].set_title("B = 0, A = 20, Alpha = 0.5, Beta = 1.5")
axs[2].set_title("B = 0, A = 1, Alpha = 10, Beta = 1.5")
axs[3].set_title("B = 0, A = 1, Alpha = 0.5, Beta = 30")

fig.suptitle('V(t) vs. t for Different Model Parameters', weight='bold', size=16)
plt.xlabel('t', size=16, weight='bold')
plt.ylabel('V(t)', size=16, weight='bold')

fig.set_size_inches(10, 10)
plt.show()
```

```python
hiv_data = np.loadtxt("HIVseries.csv", delimiter=",")
time, viral_load_data = hiv_data.T
print("Time:", time)
plt.plot(time, viral_load_data, '+')

viral_load2 = 9000 * np.exp(-(1/3650)*time) + 170000 * np.exp(-0.6*time)
plt.plot(time, viral_load2, 'r')

plt.xlabel('Time')
plt.ylabel('Virtual Load')
plt.title('HIV Series Data')
plt.show()
```

Analysis:
(Note ~ The lowest two data points around t = 0 were ignored, as discussed in class.)
To fit a curve to the data, I first set alpha to 1/3650, as we discussed in class. Then I adjusted beta, which we know must be larger than alpha. Adjusting beta effectively meant adjusting how deeply the curve bent (exponential decay). Once beta was adjusted properly, I then adjusted the coefficients A and B. A + B should be the value of V(t=0), so the sum of these two helped me to set the left part of the curve to the correct height. Then I adjusted B and A, keeping their sum constant, to finally fit the entire curve!

The inverse of the t-cell infection rate (1/alpha) is 3650. This is significantly larger than the latency period of HIV, which is just 10 years. 

```python
A = 1
tau = 1
t = np.arange(0, 2.1, 0.1)
W = A * (np.exp(-t/tau) - 1 + (t/tau))
plt.plot(t, W)
plt.title('W(t) when A = 1, Tau = 1', weight='bold', size=18)
plt.xlabel('t', weight='bold', size=14)
plt.ylabel('W(t)', weight='bold', size=14)
plt.show()

W1 = 1 * (np.exp(-t/1) - 1 + (t/1))
W2 = 1 * (np.exp(-t/0.5) - 1 + (t/0.5))
W3 = 1 * (np.exp(-t/2) - 1 + (t/2))
W4 = 0.5 * (np.exp(-t/1) - 1 + (t/1))
W5 = 0.5 * (np.exp(-t/0.5) - 1 + (t/0.5))
W6 = 0.5 * (np.exp(-t/2) - 1 + (t/2))
W7 = 2 * (np.exp(-t/1) - 1 + (t/1))
W8 = 2 * (np.exp(-t/0.5) - 1 + (t/0.5))
W9 = 2 * (np.exp(-t/2) - 1 + (t/2))

plt.plot(t, W1, 'bx')
plt.plot(t, W2, 'g+')
plt.plot(t, W3, 'r--')
plt.plot(t, W4, 'c-')
plt.plot(t, W5, 'm:')
plt.plot(t, W6, 'y-.')
plt.plot(t, W7, 'k.')
plt.plot(t, W8, color='purple', marker='s', linewidth=0)
plt.plot(t, W9, color='orange', marker='D', linewidth=0)

ax = plt.gca()
lines = ax.get_lines()
for i in range(9):
    lines[i].set_label("Curve " + str(i + 1))

plt.legend()
plt.title('W(t) for Different Values of A, Tau', weight='bold', size=18)
plt.xlabel('t', weight='bold', size=14)
plt.ylabel('W(t)', weight='bold', size=14)
plt.show()
```

Analysis:
This was a straightforward problem in the chapter. The first plot shows W(t) for values of t from 0 to 2, when the parameters A and Tau are both set to 1. The second plot shows W(t) for values of t from 0 to 2, when the parameters A and Tau are adjusted to double and half their values in different combinations (see the code for this). I then 'decorated' the different curves, as the reading suggested. 

```python
dataA = np.loadtxt("g149novickA.csv", delimiter=",")
t, V_t = dataA.T
plt.plot(t, V_t, 'bo')
plt.xlabel('t', weight='bold', size=14)
plt.ylabel('V(t)', weight='bold', size=14)

V = 1 - np.exp(-t/3.5)
plt.plot(t, V)

V1 = 1 - np.exp(-t/5)
plt.plot(t, V1)

V2 = 1 - np.exp(-t/8)
plt.plot(t, V2)

V3 = 1 - np.exp(-t/2)
plt.plot(t, V3)

V4 = 1 - np.exp(-t/1)
plt.plot(t, V4)

ax = plt.gca()
lines = ax.get_lines()
lines[0].set_label("Data")
lines[1].set_label("Tau = 3.5")
lines[2].set_label("Tau = 5")
lines[3].set_label("Tau = 8")
lines[4].set_label("Tau = 2")
lines[5].set_label("Tau = 1")
plt.legend()
plt.show()
```

Analysis:
The original data from the file is shown in the above graph in blue dots. The best fitting line for this data is the V(t) equation with Tau set to 3.5. I determined this with trial and error because for larger values of Tau (>3.5), the curve lies below the data, i.e. the values of V(t) are too small to accurately capture the data. Likewise, for smaller values of Tau (<3.5), the curve lies above the data, i.e. the values of V(t) are too large to accurately capture the data. All of this is shown in the legend in the above graph. 

```python
dataB = np.loadtxt("g149novickB_reduced.csv", delimiter=",")
t, W_t = dataB.T
plt.plot(t, W_t, 'go')
plt.xlabel('t', weight='bold', size=14)
plt.ylabel('W(t)', weight='bold', size=14)

W = 1 * (np.exp(-t/18) - 1 + (t/18))
plt.plot(t, W)

W1 = 2 * (np.exp(-t/18) - 1 + (t/18))
plt.plot(t, W1)

W2 = 0.5 * (np.exp(-t/18) - 1 + (t/18))
plt.plot(t, W2)

W3 = 1 * (np.exp(-t/36) - 1 + (t/36))
plt.plot(t, W3) 

W4 = 1 * (np.exp(-t/9) - 1 + (t/9))
plt.plot(t, W4)

ax = plt.gca()
lines = ax.get_lines()
lines[0].set_label("Data")
lines[1].set_label("A = 1, Tau = 18")
lines[2].set_label("A = 2, Tau = 18")
lines[3].set_label("A = 0.5, Tau = 18")
lines[4].set_label("A = 1, Tau = 36")
lines[5].set_label("A = 1, Tau = 9")
plt.legend()
plt.show()
```

Analysis:
The original data from the file is shown in the above graph in green dots. The best fitting line for this data is the W(t) equation with A set to 1 and Tau set to 18. Again, I determined this with trial and error, although it was much more challenging because there were two variables to adjust here. Ultimately, I found that if we hold Tau constant at 18, then when A > 1, such as A = 2, the W(t) output is too large, making the curve lie too far above the data. When A < 1, such as A = 0.5, the W(t) output is too small, making the curve lie too far below the data. Additionally, when we hold A constant at 1, then when Tau > 18, such as Tau = 36, the W(t) output is too small, making the curve lie too far below the data. When Tau < 18, such as Tau = 9, the W(t) output is too large, making the curve lie too far above the data. In comparing the effects of altering A and Tau separately, it is clear that changing Tau results in more extreme differences than changing A. All of this is shown in the legend in the above graph.
